# **ĐỒ ÁN MÔN HỌC**

## **PHÂN LOẠI CHÓ MÈO**

**Thành viên:** Tô thanh Tiến - 17521130

### **Đặt vấn đề:**




*   Hiện nay, bài toán phân loại là một bài toán khá phổ biến và được áp dụng trong nhiều vấn đề.
*   Bài toán phân lớp chó mèo là một bài toán cơ sở cũng như tiền đề để áp dụng và cải tiến lên nhiều vấn đề mới và hữu dụng hơn.



### **Giới thiệu đề tài:**



*   Input: ảnh chó hoặc mèo
*   Output: phân loại chó hay mèo



### **Mô tả dữ liệu:**



*   Bộ dữ liệu được lấy ở trang kaggle: https://www.kaggle.com/tongpython/cat-and-dog bộ dữ liệu này là của SchubertSlySchubert

*   Bộ train gồm 4001 ảnh mèo và 4006 ảnh chó.
*   Bộ test gồm 1012 ảnh mèo và 1013 ảnh chó.




### **Xử lý dữ liệu và chọn model:**

**Xử lý dữ liệu:**


*   Chuyển ảnh màu RGB về ảnh xám
*   Resize bức ảnh về chung một kích thước duy nhất 64x64(logistic regression), 224x224(CNN)
*   Trích xuất đặc trưng cho bức ảnh



**Model:** Dùng CNN-Dl-VGG16, KNN và logistic regression và so sánh các kết quả với nhau.

Đánh giá kết quả thông qua 1 vài độ đo như accurancy, confusion matrix..., so sánh kết quả đầu ra của các thuật toán.
